```python
import openpyxl
import pandas as pd
```

```python
path = '지역건강통계' # 현재 파일 위치

import os

#파일 경로 위치 알아보기
def allfiles(path):
    global res
    global rootpath
    res = []
    for root, dirs, files in os.walk(path):
        rootpath = os.path.join(os.path.abspath(path))
        print(rootpath)

        for file in files:
            filepath = os.path.join(rootpath, file)
            res.append(filepath)
            
    return res

allfiles(path)
```

```python
for file in res:
    
    # 폴더 생성 (엑셀 파일 이름으로 폴더 생성하기)
    folder_dir = file[:-5]
    os.makedirs(folder_dir, exist_ok = True)
    
    for res_name in res:            
        wb = openpyxl.load_workbook(res_name)
        ws_name = wb.sheetnames # 각 엑셀 파일의 시트 이름 
        
        for sheet in ws_name:
            # pd.read_excel('읽을파일',sheet_name = '읽을 시트 이름', engine='openpyxl)
            
            try:
                df_sheet_multi = pd.read_excel(file,sheet_name = sheet.format(ws_name),engine='openpyxl')
                file_name = file.split('\\')
                name = file_name[6][12:-5] # 폴더 이름만 가져오기
                file_name = file_name[5][:8]
                print(file_name + '_' + sheet)
                #  to_csv('저장할 파일명(경로/파일명).csv') 로 저장
                df_sheet_multi.to_csv("{}\\{}_{}.csv".format(folder_dir,name,sheet))
                print("save : "+ "{}_{}.csv".format(folder_dir,sheet))
            except :
                pass
                print("Error: not found.")
```
